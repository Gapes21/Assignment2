```python
import cv2 as cv
import os
from tqdm import tqdm
from sklearn.cluster import KMeans
import numpy as np
import torch
```

```python
class FeatureBag:
    def __init__(self, directory, k = 100):
        self.k = k
        self.features = []
        for f in tqdm(os.listdir(directory)):
            img = cv.imread(os.path.join(directory, f))
            gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
            
            sift = cv.SIFT_create()
            keypoints, descriptors = sift.detectAndCompute(gray, None)
            self.features.extend(descriptors)
        print(len(self.features))
        self.cluster()
        
    def cluster(self):
        self.kmeans = KMeans(
            n_clusters = self.k,
            random_state = 42,
            verbose = True
        ).fit(self.features)
        
    def assignLabel(self, image):
        gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
        sift = cv.SIFT_create()
        k, descriptors  = sift.detectAndCompute(gray, None)
        _, histogram = np.histogram(self.kmeans.predict(descriptors.reshape(1, -1), bins = self.k, range = (0, self.k)))
        return histogram
```
